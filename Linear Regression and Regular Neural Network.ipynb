## Import Libraries

```python
import os
import h5py
import numpy as np
import pandas as pd
from pandas import DataFrame
#import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
```

## Loading Data

```python
# Load data
filename = 'N-CMAPSS_DS02.h5'
with h5py.File(filename, 'r') as hdf:
            # Development set
            W_dev = np.array(hdf.get('W_dev'))             # W
            X_s_dev = np.array(hdf.get('X_s_dev'))         # X_s
            #X_v_dev = np.array(hdf.get('X_v_dev'))         # X_v
            #T_dev = np.array(hdf.get('T_dev'))             # T
            Y_dev = np.array(hdf.get('Y_dev'))             # RUL  
            A_dev = np.array(hdf.get('A_dev'))             # Auxiliary
            
            # Test set
            W_test = np.array(hdf.get('W_test'))           # W
            X_s_test = np.array(hdf.get('X_s_test'))       # X_s
            #X_v_test = np.array(hdf.get('X_v_test'))       # X_v
            #T_test = np.array(hdf.get('T_test'))           # T
            Y_test = np.array(hdf.get('Y_test'))           # RUL  
            A_test = np.array(hdf.get('A_test'))           # Auxiliary
            
            # Varnams
            W_var = np.array(hdf.get('W_var'))
            X_s_var = np.array(hdf.get('X_s_var'))  
            #X_v_var = np.array(hdf.get('X_v_var')) 
            #T_var = np.array(hdf.get('T_var'))
            A_var = np.array(hdf.get('A_var'))
            
            # from np.array to list dtype U4/U5
            W_var = list(np.array(W_var, dtype='U20'))
            X_s_var = list(np.array(X_s_var, dtype='U20'))  
            #X_v_var = list(np.array(X_v_var, dtype='U20')) 
            #T_var = list(np.array(T_var, dtype='U20'))
            A_var = list(np.array(A_var, dtype='U20'))
```

```python
# Concat all development data
Dev = np.concatenate((W_dev, X_s_dev, A_dev, Y_dev), axis=1)

# Concat all test data
Test = np.concatenate((W_test, X_s_test, A_test, Y_test), axis=1)
```

```python
print(X_s_dev.shape)

df_Dev_cols = ['alt', 'Mach', 'TRA', 'T2', 'T24', 'T30', 'T48', 'T50', 'P15', 'P2', 'P21', 'P24',
 'Ps30', 'P40', 'P50', 'Nf', 'Nc', 'Wf', 'unit', 'cycle', 'Fc', 'hs', 'RUL']

#df_Dev_cols = ['alt', 'Mach', 'TRA', 'T2', 'Wf', 'Nf', 'Nc', 'T24', 'T30', 'T48', 'T50', 'P15', 'P2', 'P21', 'P24', 'Ps30', 'P40', 'P50', 'unit', 'cycle', 'Fc', 'hs','RUL']

df_Dev = pd.DataFrame(Dev,columns=df_Dev_cols)
df_Test = pd.DataFrame(Test,columns=df_Dev_cols)

display(df_Dev)
display(df_Test)
```

### Choosing Specific Unitis for Training and Testing

```python
df_Dev = df_Dev[(df_Dev['unit'] == 16) | (df_Dev['unit'] == 18) | (df_Dev['unit'] == 20)]
df_Test = df_Test[(df_Test['unit'] == 11) | (df_Test['unit'] == 14) | (df_Test['unit'] == 15)]
#df_Test = df_Test[(df_Test['unit'] == 14)]
```

```python
print(np.shape(df_Dev))
print(np.shape(df_Test))
```

### Preprocessing

```python
def preprocess_inputs(df):
    df = df.copy()  
    
    y = df[['RUL', 'unit', 'cycle']]
    X = df.drop('RUL', axis=1).copy()
    
    return X, y
```

```python
x_train, y_train = preprocess_inputs(df_Dev)
x_test, y_test = preprocess_inputs(df_Test)
```

## Units and Cycle

Columns "unit" and "cycle" are not interesting for linear regression or NN.

```python
x_train.drop("unit",axis=1,inplace=True)
x_test.drop("unit",axis=1,inplace=True)
#y_train.drop("unit",axis=1,inplace=True)
#y_test.drop("unit",axis=1,inplace=True)

x_train.drop("cycle",axis=1,inplace=True)
x_test.drop("cycle",axis=1,inplace=True)
#y_train.drop("cycle",axis=1,inplace=True)
#y_test.drop("cycle",axis=1,inplace=True)

x_train.drop("Fc",axis=1,inplace=True)
x_test.drop("Fc",axis=1,inplace=True)

x_train.drop("hs",axis=1,inplace=True)
x_test.drop("hs",axis=1,inplace=True)

y_train = y_train.RUL
y_test = y_test.RUL

display(x_train)
display(y_train)
display(x_test)
display(y_test)
```

### Normalization

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)

#from sklearn.decomposition import PCA
# Make an instance of the Model
#pca = PCA(.98)
#pca.fit(X_train)
#X_train = pca.transform(X_train)
#X_test = pca.transform(X_test)
```

### Machine Learning or Deep Learning Models

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
y_pred_reg = reg.predict(x_test)

from sklearn.metrics import mean_squared_error

rmse = mean_squared_error(y_pred_reg,y_test, squared=False)
print('RMSE = ', rmse)

print('Actual RUL:')
display(pd.DataFrame(y_test, columns = ['RUL']))
print('Predicted RUL:')
display(pd.DataFrame(y_pred_reg, columns = ['RUL']))


#plt.figure(figsize=(10, 6))
#plt.title("Predicted vs. Actual :)", size=16)
#plt.scatter(range(1253743), np.squeeze(y_test))
#plt.plot(range(1253743), np.squeeze(y_pred_reg), c="red")
#plt.show()
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(32, input_dim=18, kernel_initializer='normal', activation='relu'))
model.add(tf.keras.layers.Dropout(0.12))
model.add(tf.keras.layers.Dense(32, kernel_initializer='normal', activation='relu'))
model.add(tf.keras.layers.Dense(32, kernel_initializer='normal', activation='relu'))
model.add(tf.keras.layers.Dense(1, kernel_initializer='normal', activation='linear'))
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mean_squared_error'])
```

```python
history = model.fit(
    x_train,
    y_train,
    validation_split=0.1,
    batch_size=512,
    epochs=2500,
    callbacks=[
        tf.keras.callbacks.EarlyStopping(
            monitor='val_loss',
            patience=7,
            restore_best_weights=True
        )
    ]
)
```

```python
model.save('ANN.h5')
```

```python
y_pred_ann = model.predict(x_test)

rmse = mean_squared_error(y_pred_ann,y_test, squared=False)
print('RMSE = ', rmse)
```
